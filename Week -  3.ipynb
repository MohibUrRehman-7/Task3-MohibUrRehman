```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("raw_skills.csv")
df
```

```python
vectorizer = TfidfVectorizer(lowercase=True)
role_vectors = vectorizer.fit_transform(df["skills"])

print("Vocabulary size:", len(vectorizer.vocabulary_))
print(vectorizer.get_feature_names_out())
```

```python
num_skills = 3

user_skills = []
print(f"Enter at least {num_skills} skills, one at a time.")
for i in range(num_skills):
    skill = input(f"Skill {i+1}: ")
    user_skills.append(skill)

print("\nYour entered skills:", user_skills)

user_text = " ".join(user_skills)
user_vector = vectorizer.transform([user_text])
```

```python
scores = cosine_similarity(user_vector, role_vectors).flatten()
df["match_score"] = scores
df[["role", "match_score"]]
```

```python
top_matches = df.sort_values("match_score", ascending=False).head(3)
top_matches[["role", "match_score"]]
```

```python
print("Your skills:", user_skills)
print("\nTop", 3, "recommended career paths:\n")
for i, row in enumerate(top_matches.itertuples(), start=1):
    print(f"{i}. {row.role}  (match score: {row.match_score:.2f})")
```
